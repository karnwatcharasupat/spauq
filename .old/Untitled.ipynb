```python
import numpy as np
from scipy import signal as ss
```

```python
x = np.random.randn(100)
```

```python
y = np.roll(x, -25)
y[-25:] = np.random.randn(25)
```

```python
y.shape
```

```python
ss.correlation_lags(100, 100)[np.argmax(ss.correlate(x, y))]
```

```python
from matplotlib import pyplot as plt
```

```python
plt.plot(x)
plt.plot(np.roll(y, 25))
```
